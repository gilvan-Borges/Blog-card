```python
# Importando bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# 1) Exibir uma série que indique o número de alunos por titulação
Titulação = pd.Series(
    ['Doutorado', 'Mestrado', 'Especialização', 'Graduação', 'Cursos Técnicos'],
    index=[20, 15, 85, 145, 320]
)
print("Série de alunos por titulação:")
print(Titulação)

# -------------------------------
# 2) Exibir uma tabela que indique o número de alunos por titulação
df_titulacao = pd.DataFrame({
    'Titulação': ['Doutorado', 'Mestrado', 'Especialização', 'Graduação', 'Cursos Técnicos'],
    'Quantidade': [20, 15, 85, 145, 320]
})
print("\nTabela de alunos por titulação:")
print(df_titulacao)

# -------------------------------
# 3) Tabela com calorias e percentual de gordura
df_alimentos = pd.DataFrame({
    'calorias': [200, 350, 550],
    'gordura (%)': [0, 15, 35]
}, index=['banana', 'macarrão', 'cachorro quente'])
print("\nTabela de calorias e gordura:")
print(df_alimentos)

# -------------------------------
# 4) Criar uma matriz com 50 elementos e 3 dimensões
v = np.array(range(50)).reshape(2, 5, 5)
print("\nMatriz com 50 elementos e 3 dimensões:")
print("Shape = ", v.shape)
print("Número de dimensões = ", v.ndim)
print("Número de elementos = ", v.size)
print("Tensor v = \n", v)

# -------------------------------
# 5) Criar um gráfico de pizza com 3 itens, destacando 1
fatias = [8, 12, 10]
atividades = ['Lazer', 'Estudo', 'Viagem']
colunas = ['r', 'm', 'y']

plt.pie(fatias, labels=atividades, colors=colunas, startangle=90, shadow=True,
        explode=(0, 0.1, 0))
plt.title("Distribuição de atividades")
plt.show()

# -------------------------------
# 6) Criar um gráfico tipo linha
x = [1, 3, 5]
y = [1, 2, 5]

plt.plot(x, y, label='Uma legenda')
plt.title('Exemplo utilizando Plot')
plt.xlabel('Variável 1')
plt.ylabel('Variável 2')
plt.legend()
plt.show()

# -------------------------------
# 7) Criar um gráfico com pontos em formato de estrela
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [5, 2, 4, 5, 6, 8, 4, 8]

plt.scatter(x, y, label='Pontos', color='b', marker='*', s=100)
plt.title("Gráfico de pontos em formato de estrela")
plt.legend()
plt.show()
```
